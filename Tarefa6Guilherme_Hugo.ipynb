```python
##Este notebook realiza as operações para calcular a população de 1994 até 2022 
##e representa os resultado no gráfico

import matplotlib.pyplot as plt ##Importa a biblioteca mathplot para o grafico
import numpy as np ##Biblioteca numpy para realiazar contas matematicas.

a = np.linspace (1994,2022,28); ##valores que representam os anos, com acrescimo de 1 no eixo x
e = (a-1994)/40; ##Potencia da formula para o calculo da população
p = 800000*np.exp(e); ##Formula para referente ao crescimento da população
plt.figure();
plt.plot(a,p);
plt.xlabel('Ano'); ##Eixo X
plt.ylabel('População'); ##Eixo Y
plt.title('População entre os anos de 1994 e 2022') ##Titulo
plt.show();
```

```python
##Este notebook transforma a temperatura de Fahreint pra Celsisus e representa 
##no gráfico mostrando sua equivalencia.

import matplotlib.pyplot as plt ##Importa a biblioteca mathplot para o grafico
import numpy as np ##Biblioteca numpy para realiazar contas matematicas.

##Gráfico mostra a relação entre os valores de temperatura de Celsius para  Fahreint

c = np.linspace(-20,100,100); ##Valores em Celsius, de -20 até 100, com 100 termos
f = 9/5*c+32; ##Formula para passar de Celsius para Fahreint

plt.figure();
plt.plot(c,f);
plt.xlabel('Celsius (°c)') ##Nome Eixo X
plt.ylabel('Fahreint (°f)') ##Nome Eixo Y
plt.title('Temperatura equivalente de Celsius para Fahreint') ##Titulo
plt.show();
```
