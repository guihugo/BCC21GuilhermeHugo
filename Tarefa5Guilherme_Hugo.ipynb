```python
##Guilherme Hugo de Araujo Silva RA 11202230757
##Calcula o gráfico da altura da bola de acordo com o tempo

import matplotlib.pyplot as plt ##Importa a biblioteca mathplot para o grafico
import numpy as np ##Biblioteca numpy para realiazar contas matematicas.

t = np.linspace(0,5,50); ##Define os valores de t, os valores do tempo
h = -t**2+4*t; ##Função da altura em relação ao tempo

plt.figure() 
plt.plot(t, h)
plt.xlabel('tempo (s)') ##altera o nome do vetor x
plt.ylabel('altura (m)') ##altera o nome do vetor y
plt.title('Altura da Bola') #altera o nome do titulo
plt.show()
maior=max(h); ##calcula o maior valor para a função h
print(f"A maior altura é igual a: {maior:.3f}");
```
