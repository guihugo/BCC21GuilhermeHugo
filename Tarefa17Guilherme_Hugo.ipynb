```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

jogos = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')

corr_publico_renda = jogos['Público'].corr(jogos['Renda (R$)']) ##Calcula a correlação

plt.figure(figsize = (8,8)) ##Plot
plt.plot(jogos['Público'], jogos['Renda (R$)'], marker = 'o',color = 'red', linestyle ='')
plt.title('Renda')
plt.xlabel('Renda')
plt.ylabel('Público')
plt.show()

print (f'A correlação entre Público e renda é de: {corr_publico_renda:.2f}')
```
