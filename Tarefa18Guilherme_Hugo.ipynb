```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('https://nbviewer.org/github/BMClab/BasesComputacionais/blob/master/dados/BDSinfo.csv', sep='\t')

mediaX = np.mean(df['Height']) ##Calcula a média do eixo x
mediaY = np.mean(df['FootLen']) ##Calcula a media do eixo y
desvioX = df['Height'] - mediaX 
desvioY = df['FootLen'] - mediaY
m = np.sum(desvioX*desvioY)/np.sum(desvioX**2)
b = mediaY - m*mediaX ##Formúla dado para calculo de regressão
p = df['FootLen'].corr( df['Height'] ) ##Correlação

plt.figure(figsize=(7,7))
plt.plot(df['Height'], df['FootLen'], marker = 'o',color = 'red', linestyle ='', label = f'Correlação: {p:.2f}')
plt.plot(df['Height'], m*df['Height']+b, color ='blue')
plt.title('Correlação entre Pé e Altura')
plt.xlabel ('Altura (Cm)')
plt.ylabel ('Tamanho do Pé (Cm)')
plt.legend()
plt.show()


print(f'A correlação é de: {p:.2f}')

```
