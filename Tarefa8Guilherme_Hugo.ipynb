```python
import matplotlib.pyplot as plt ##Importa a biblioteca mathplot para o grafico
import numpy as np ##Biblioteca numpy para realiazar contas matematicas.

fig, ax = plt.subplots(figsize = (9, 6)) ##Aumenta o tamanho do gráfico

x = np.arange(-10,10,0.001); ##Define o tamanho do array
y = x**2-np.sin(0.784*x**2)-2; ##Fórmula da Equação
raizes = x[np.isclose(y,0, atol=0.002)]; ##Define a equação

plt.plot(x, y) ##Forma o gráfico
plt.xlabel('x') ##Eixo x
plt.ylabel('f(x)') ##Eixo Y
plt.grid()
plt.show()
print('As raízes da função são:', raizes) ##Print dos valores das raízes
```

```python
import matplotlib.pyplot as plt ##Importa a biblioteca mathplot para o grafico
import numpy as np ##Biblioteca numpy para realiazar contas matematicas.


fig, ax = plt.subplots(figsize = (9, 6)) ##Aumenta o tamanho do gráfico

x = np.arange(-10, 11); ##Define o array do valor x
f = abs(x - 2) + abs(2*x+1) - x - 6; ##Equação dada pelo exercício

plt.plot(x, f) 
plt.xlabel('x') ##Eixo x
plt.ylabel('f(x)') ##Eixo Y
plt.grid()
plt.show()


print('O valor mínimo é igual a ',np.min(x[(x>0) & (f>0)])); ##Print do menor
                                                ##valor para x > 0 e f(x) > 0

```

```python
x = [53, 83,64] ##Peso em Kg

y = [1.58, 1.75,1.79] ##Altura em M

for a, b in zip(x, y): ##loop para calculo de IMC
  i = a/b**2 ##Fórmula IMC

  if i < 17: ##Verificação do valor de IMC
      print("Muito abaixo do peso")
  else:
      if 17 < i < 18.5:
        print("Abaixo do peso")
      else:
        if 18.5 < i < 25:
          print("Peso Normal")
        else:
          if 25 < i < 30:
            print("Acima do peso")
          else:
            if 30 < i < 35:
              print("Obesidade Grau I")  
            else:
              if 35 < i < 40:
                print("Obesidade Grau II")
              else: 
                if i > 40:
                  print("Obesidade Grau III")        

```
