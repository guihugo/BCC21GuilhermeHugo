```python
import pandas as pd ##Importa para utilizar banco de dados
import matplotlib.pyplot as plt ##Importa a biblioteca mathplot para o grafico


analfabeta = pd.read_csv ('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/analfabetismo.csv')
##link para importar banco de dados de pessoas analfabetas

plt.figure(figsize=(7,7)) ##Altera a proporção do gráfico
plt.plot(analfabeta['Ano'], analfabeta['São Paulo']) ##Descreve qual dados tem que ser buscado
plt.title('Analfabetismo em São Paulo')
plt.xlabel('Ano') ##Eixo X
plt.ylabel('População (10⁶)') ##Eixo Y
plt.show()
```
