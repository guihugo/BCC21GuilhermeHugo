```python
import pandas as pd
import matplotlib.pyplot as plt

inflacao = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflacaoMensal.csv')

nova = inflacao.query("Ano >= 1995")

mediana = nova.query("Mês == 3")['Inflação'].median()

media = nova.query("Mês == 3")['Inflação'].mean()

dp = nova.query("Mês == 3")['Inflação'].std()

print (f'A mediana da inflação a partir de 1995 é de {mediana}')
print (f'A media da inflação a partir de 1995 é de {media:.2f}')
print (f'O desvio da inflação a partie de 1995 é de {dp:.3f}')

##A diferença na média é em razão da quantidade de meses pela qual a inflação é divida,
##e a mediana ser menor é em razão do periodo ser menor, e o meio ser num periodo de menor inflação


##Gráfico do Histograma
plt.figure(figsize=(8,9))
plt.hist([nova.query("Mês == 3")['Inflação']])
plt.xlabel('Inflação');
plt.ylabel('Mês')
plt.show()

```

```python
##Tarefa 2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ipca = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflacaoMensal.csv')

mediana= ipca.query("Mês == 3")['Inflação'].median()
media = ipca.query("Mês == 3")['Inflação'].mean()
dp = ipca.query("Mês == 3")['Inflação'].std()

print (f'A mediana da inflação é de {mediana:.2f}')
print (f'A media da inflação é de {media:.2f}')
print (f'O desvio padrão é de {dp:.2f}')

plt.figure(figsize=(8,9))
plt.hist([ipca.query("Mês == 3")['Inflação']])
plt.xlabel('Mês');
plt.ylabel('Inflação')
plt.show()
```
